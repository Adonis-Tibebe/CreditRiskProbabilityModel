## Visualization of key features of dataset 
- Distribution of **numeric columns** of the dataset
- Corelation Analysis
- Outlier Detection

```python
# Visualize distributions of numerical features
numerical_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, (len(numerical_cols) + 1) // 2, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# lets look at the distribution of amount and value in a bit more detail
#for value column
plt.figure(figsize=(15, 8))
sns.histplot(df["Value"],bins=40 , kde=True)
plt.title(f'Distribution of {df["Value"].name}')
    
plt.tight_layout()
plt.show()

```

```python
# visualization of distribution of Amount, negative and positive values included
plt.figure(figsize=(15, 8))
sns.histplot(df[df["Amount"] > 0]["Amount"], color="green",bins=100, label="Positive", kde=True)
sns.histplot(df[df["Amount"] < 0]["Amount"], color="red",bins=100, label="Negative", kde=True)
plt.legend()
plt.xlabel("log")
plt.title(f'Distribution of Positive and Negative {col} Values')
plt.tight_layout()
plt.show()
```

### Interpretation of Amount and Value Distributions

The plot above shows the distribution of the **Amount** column, with positive values highlighted in green and negative values in red. Here’s what we can interpret:

- **Positive Amounts (Green):**  
  The majority of transactions have positive amounts, indicating standard inflow or credit transactions. The distribution is highly skewed to the right, with most values clustered near zero and a long tail of larger amounts.

- **Negative Amounts (Red):**  
  There is a significant presence of negative values, which likely represent outflows, refunds, or chargebacks. These are also concentrated near zero but are distinctly separated from the positive values.

- **Skewness and Outliers:**  
  Both positive and negative distributions are heavily skewed, suggesting the presence of a few very large transactions (outliers) compared to the majority.

For the **Value** column, the distribution is very similar to the positive side of the Amount column but does **not** include negative values.The Values column possibly representing the absolute values of the Amount column.

```python
# Visualize distributions of Key categorical features
categorical_cols = ["ProviderId", "ProductId","ProductCategory","ChannelId" ,"PricingStrategy"]

plt.figure(figsize=(16, 8))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(2, (len(categorical_cols) + 1) // 2, i)
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Countplot of {col}')
plt.tight_layout()
plt.show()
```

### Distribution Analysis

The above code visualizes the distributions of key numerical and categorical features in the dataset.  
- **Numerical features**: Histograms with KDE curves help us understand the spread, skewness, and presence of outliers.
- **Categorical features**: Bar plots show the frequency of each(does not include catagorical columns with too many unique values) category, helping to spot imbalances or rare categories.

These visualizations are essential for understanding the underlying structure of your data and for identifying potential issues before modeling.

```python
# Correlation Analysis: Visualize correlations between numerical features

# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)
corr_matrix.drop(columns=['CountryCode'], inplace=True)  # Drop datetime column if present

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

## The above shows the corelation matrix heat map of our dataset

```python
# Outlier Detection for Amount and Value columns

plt.figure(figsize=(14, 6))

# Boxplot for Amount
plt.subplot(1, 2, 1)
sns.boxplot(x=df["Amount"])
plt.title("Boxplot of Amount")
plt.xlabel("Amount")

# Boxplot for Value
plt.subplot(1, 2, 2)
sns.boxplot(x=df["Value"])
plt.title("Boxplot of Value")
plt.xlabel("Value")

plt.tight_layout()
plt.show()
```

### Outlier Detection for Amount and Value

The boxplots above clearly show the presence of numerous outliers in both the **Amount** and **Value** columns. Most data points are concentrated near zero(looks so since the numbers are in logarithmic format for normalization since there range is quite large), while a significant number of extreme values (outliers) are spread far from the center, on both the positive and negative sides for Amount, and only on the positive side for Value.

This observation is consistent with what was already predicted in the **summary statistics** section of the notebook, where the large difference between the mean and median, as well as the high standard deviation, indicated the presence of extreme values. It also aligns with the earlier **distribution plots**, which showed highly skewed distributions and long tails for both columns.

These outliers may have a substantial impact on further analysis and modeling, so it is important to consider appropriate handling strategies, such as transformation, capping
